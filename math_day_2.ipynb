```python
#outliers
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define data set
dataset = [11, 10, 12, 14, 12, 15, 14, 13, 15, 102, 12, 14, 17, 19, 107, 10, 13, 12, 14, 12, 108, 12, 11, 14, 13, 15, 10, 15, 12, 10, 14, 13, 15, 10]
```

**Z score** (a way of finding outliers)

measures how many standard deviations a specific data point is from the dataset's mean

Z score = X - μ / σ

Where:

x: The observed data point

μ: The mean of the dataset

σ: The standard deviation of the datase

```python
## Z score (a way of finding outliers)

outliers = []
def detect_outliers(data):
  threshold = 3          ## 3 STD deviations
  mean = np.mean(data)
  std = np.std(data)

  for i in data:
    z_score = (i-mean)/std   # Z score formula
    if np.abs(z_score) > threshold:   # checking the condition for 3 STD deviations
      outliers.append(i)      #appneding/storing the data into outliers = []
  return outliers
```

```python
detect_outliers(dataset)
```

**IQR** (Interquartile Range)

IQR tells us the range where the bulk of the values lie

IQR = Q3 - Q1

IQR we find-out for **lower fences** and **higher fences**

the upper and lower fences are boundaries used to identify outliers in a dataset

**lower fences** = Q1 - 1.5 (IQR)

**higher fences** = Q3 + 1.5 (IQR)

```python
## IQR (Interquartile Range)
# 1. Sort the data
# 2. calculate Q1 and Q3
# 3. calculate IQR (Q3 - Q1)
# 4. find lower fence = Q1 - 1.5 (IQR)
# 5. find higher fence = Q3 + 1.5 (IQR)
```

```python
#step 1 sort the data

dataset = sorted(dataset)
dataset
```

```python
#2. finding Q1, Q3

q1, q3 = np.percentile(dataset, [25, 75])
print(q1, q3)
```

```python
#3. IQR (Q3 - Q1)
iqr = q3 - q1
print(iqr)
```

```python
#4. find the lower fence and higher fence
lower_fence = q1 - (1.5 * iqr)
higher_fence = q3 + (1.5 * iqr)
print(lower_fence, higher_fence)
```
