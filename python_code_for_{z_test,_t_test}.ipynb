**Z-Test:**

suppose the IQ in a certain population is normal distribution with a mean of μ = 100 and standard deviation of σ (sigma) = 15
a researcher wants to know if a new drug affects IQ levels, so he recurites 20 patients to try it and record their IQ levels


The following code shows how to perform **Z-test** in python and determine if new drug causes significant different IQ level


```python
from statsmodels.stats.weightstats import ztest as ztest

#Enter IQ levels from 20 patients
data = [88, 92, 94, 94, 96, 97, 97, 97, 99, 99, 105, 109, 109, 109, 110, 112, 112, 112, 114, 115]
```

```python
#ztest
z_stat, p_value = ztest(data, value=100)

print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")
```

we reject the null hypothesis H0

**z-test** = 1.582295902196661 > **P-value**: 0.11358203578598267

**Internl**

1. **Problem Statement**
   "Does a new drug affect IQ?"

                   ↓

2. Collect Data
   20 patients, IQ scores recorded

        
                  ↓
3. Sample Statistics
   Mean = 100.4, Std Dev = 8.39

                  ↓
4. Set Up Hypotheses
   H₀: μ = 100 (No effect)
   H₁: μ ≠ 100 (Does affect)

                  ↓
5. Choose Alpha
   α = 0.05

                   ↓
6. Choose Test
   Z-test (population σ known)

                  ↓
7. Calculate Test Statistic
   Z = 1.5823

                 ↓
8. Find Critical Value
   ±1.96 (for two-tailed, α = 0.05)

                  ↓
9. Compare Z to Critical Value
   1.5823 < 1.96 → FAIL TO REJECT

                 ↓
10. Calculate P-value
    P = 0.1136

                ↓
11. Compare P to Alpha
    0.1136 > 0.05 → FAIL TO REJECT

               ↓
12. Final Decision
    ❌ Drug does NOT significantly affect IQ.

**t-test**

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import math
```

```python
np.random.seed(6)
#generates a fake population (school_ages) of students with average age around 53 (18 + after getting mu= value it adds 18)
school_ages = stats.poisson.rvs(loc= 18, mu = 35, size = 1500)
classA_ages = stats.poisson.rvs(loc = 18, mu = 30, size = 60)
```

```python
classA_ages.mean()
```

```python
_,p_value = ttest_1samp(classA_ages, popmean= school_ages.mean())
p_value
```

```python
school_ages.mean()
```

```python
if p_value < 0.05:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```
